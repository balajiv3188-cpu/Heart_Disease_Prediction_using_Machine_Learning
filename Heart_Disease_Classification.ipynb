```python
df.drop_duplicates(keep='last',inplace=True)
```

```python
from sklearn.preprocessing import OrdinalEncoder

categorical_columns = ['HeartDisease', 'Smoking', 'AlcoholDrinking','Stroke','DiffWalking','Sex','AgeCategory','Race','Diabetic','PhysicalActivity','GenHealth','Asthma','KidneyDisease','SkinCancer']
a = OrdinalEncoder()

df[categorical_columns] = a.fit_transform(df[categorical_columns])
```

```python
df.info()
```

```python
df
```

```python
X = df.drop("HeartDisease",axis=1)
```

```python
Y = df["HeartDisease"]
```

```python
X
```

```python
Y
```

```python
df["HeartDisease"].value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.20)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier().fit(x_train,y_train)
```

```python
import numpy as np
balaji=np.array([[1,32,0,0,4,23.3,5.8,120,0,1,15,16,18,1,0,4,2]])
```

```python
model.predict(balaji)
```

```python
import pickle
pickle.dump(model,open("bj.pkl","wb")) #wb write to binary
```

```python
#confusion matrix
y_test.value_counts()
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
y_predict=model.predict(x_test)
y_predict
```

```python
confusion_matrix(y_test,y_predict)
```

```python
accuracy_score(y_test,y_predict)
```

```python
print(classification_report(y_test,y_predict))
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier().fit(x_train,y_train)
y_predict=model.predict(x_test)
print(classification_report(y_test,y_predict))
```

```python
from xgboost import XGBClassifier
model = XGBClassifier().fit(x_train, y_train)
y_predict = model.predict(x_test)
print(classification_report(y_test, y_predict))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model=GradientBoostingClassifier().fit(x_train,y_train)
y_predict=model.predict(x_test)
print(classification_report(y_test,y_predict))
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier().fit(x_train,y_train)
y_predict=model.predict(x_test)
print(classification_report(y_test,y_predict))
```

```python
from sklearn.ensemble import AdaBoostClassifier
model=AdaBoostClassifier().fit(x_train,y_train)
y_predict=model.predict(x_test)
print(classification_report(y_test,y_predict))
```
